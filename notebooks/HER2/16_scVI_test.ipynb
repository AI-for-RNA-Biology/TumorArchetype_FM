```python
import scanpy as sc
import scvi


adata = sc.read_h5ad("../../results/HER2/molecular/gene_embedding_HER2.h5ad")


```

```python
latent_dims = [10, 20, 30, 50]
models = {}
elbos = []

for d in latent_dims:
    scvi.model.SCVI.setup_anndata(adata, batch_key="tumor")
    model = scvi.model.SCVI(adata, n_latent=d)
    model.train()
    models[d] = model
```

```python
for d, model in models.items():
    elbos.append(model.get_elbo())
```

```python
import numpy as np

for i, d in enumerate(latent_dims):
    print(f"{d} dims → ELBO: {elbos[i]}")
```
