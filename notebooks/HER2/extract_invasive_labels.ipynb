```python
import pandas as pd
import sys
sys.path.append("../..")

from digitalhistopathology.benchmark.benchmark_utils import select_best_UMAP_and_kmeans_parameters
```

```python
import json
import os 
import glob

with open("../../config/config_notebooks_HER2.json", "r") as config_file:
    config = json.load(config_file)
```

```python
path_ = os.path.join(config["retrained_benchmark_folder1"], "invasive_cancer_clustering_strategy_silhouette_and_batch_effect", "kmeans")


for name in config["retrained_model_list1"]:
    path = f"{path_}/{name}"
    files = glob.glob(os.path.join(path, "scores_umap*7_clusters.json"))
    print(files)
    ## Retrieve the best UMAP parameters
    best_params = select_best_UMAP_and_kmeans_parameters(files=files,
                                                            strategy_silhouette_then_batch_effect=False)
    n_cluster = best_params['n_clusters']



    ## Save the labels as a DataFrame
    labels = pd.DataFrame(best_params["labels"], best_params["samples"])
    labels.columns = ["predicted_label"]
    labels['label'] = ""
    labels.to_csv(os.path.join(path, f"invasive_labels_{n_cluster}_clusters_umap_min_dist_{best_params['min_dist']}_n_neighbors_{best_params['n_neighbors']}.csv"))
```
