```python
import json 
import os
from sklearn.metrics import adjusted_rand_score

for k in range(4, 10):
    with open(os.path.join(path, f"scores_umap_across_parameters_{k}_clusters.json"), 'r') as f:
        scores = json.load(f)
    patients = [sample.split("_")[0] for sample in scores["samples"]]
    
    for min_dist, dicto in scores.items():
        if min_dist == "samples":
            continue
        for n_neighbors, metrics in dicto.items():
            ari_score = adjusted_rand_score(patients, metrics['labels'])
            metrics['ARI_patient'] = ari_score

    with open(os.path.join(path, f"scores_umap_across_parameters_{k}_clusters.json"), 'w') as f:
        json.dump(scores, f)
```
