```python
import sys
import pandas as pd

sys.path.append("../..")
from digitalhistopathology.datasets.real_datasets import VisiumHDdataset
```

```python
ovarian_dataset = VisiumHDdataset(
    patches_folder="../../results/Ovarian/compute_patches/all",
    saving_emb_folder=None, 
    name = "Ovarian",
    raw_images_dir = "/storage/research/dbmr_luisierlab/database/Ovarian_Visium_GTOP/Visium_HD",
    spaceranger_dir = "/storage/research/dbmr_luisierlab/temp/lfournier/Data/Ovarian_Visium_GTOP/hg38/spaceranger",
    spot_diameter_micron=100
)
```

```python
gtf_path="/storage/research/dbmr_luisierlab/group/genomics/annotation/hg38/GENCODE/gencode.v44.chr_patch_hapl_scaff.annotation.gtf"
```

```python
ovarian_dataset.preprocess(gtf_path)
```

```python
# Let's filter to keep only the reliably expressed genes
import numpy as np
from bulkanalysis.filtering import gmm_threshold

thresholds = []
reliably_expressed_genes_all = []
for gene_emb in ovarian_dataset.gene_embeddings:
    sample_name = gene_emb.obs.iloc[0]["source_file"]
    pseudobulk = np.array(gene_emb.X.sum(axis=0)).flatten()
    pseudobulk_log = np.log2(pseudobulk+1)
    t = gmm_threshold(pseudobulk_log[pseudobulk_log > 0],
    return_int=False,
    quantile_threshold=0.9, 
    background_quantile=True,
    title=sample_name)
    df_pseudobulk = pd.DataFrame(pseudobulk_log, index=gene_emb.var_names, columns=["expression"])
    reliably_expressed_genes = df_pseudobulk[df_pseudobulk["expression"] > t].index.tolist()
    thresholds.append(t)
    reliably_expressed_genes_all.append(reliably_expressed_genes)
```

```python
# Number of genes in the union
len(list(set(reliably_expressed_genes_all[0]) | set(reliably_expressed_genes_all[1])))
```

```python
# Number of genes in the intersection
len(list(set(reliably_expressed_genes_all[0]) & set(reliably_expressed_genes_all[1])))
```

```python
# Keep the intersection
list_common_genes = list(set(reliably_expressed_genes_all[0]) & set(reliably_expressed_genes_all[1]))
```

```python
## Concatenate the anndata
import anndata as ad

concat_ad = ad.concat(ovarian_dataset.gene_embeddings, axis=0)
```

```python


labels = pd.read_csv("../../results/Ovarian/compute_patches/all/spots_labels.csv", index_col=0)

concat_ad.obs["spot_name"] = concat_ad.obs["source_file"] + "_spot" + concat_ad.obs["x"].astype(str) + "x" + concat_ad.obs["y"].astype(str)
concat_ad.obs = concat_ad.obs.reset_index().set_index("spot_name")
```

```python
df_genes = concat_ad.to_df()
```

```python
# Keep only reliably expressed genes
df_genes = df_genes[[col for col in df_genes.columns if col in list_common_genes]]
```
